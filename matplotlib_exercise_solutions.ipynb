```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

```python
x = range(-100, 100)
y1 = [(num ** 2 - num + 2) for num in x]
```

```python
plt.plot(x, y1)

plt.title('$y1 = x^2 -x +2$')
plt.xlim(-100, 100)
plt.ylim(-10, 5000)

plt.savefig('figure1.png')
```

```python
y2 = [(num ** 0.5) for num in x]
plt.title= ('y = \sqrt{x}$')
plt.plot(x, y2)
plt.xlimit(-10, 1000)
plt.savefig('figure2.png')
```

```python
y3 = [(num ** 3) for num in x]
plt.plot(x, y3)
plt.savefig('figure3.png')
```

```python
y4 = if (num+1) != 0:
    then [1/(num+1) for num in x]
    else:
        next
plt.plot(x,y4)
plt.savefig('figure4.png')
```
