```python
# importing required libraries for seaborn
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pydataset import data

# setting styles for seaborn
sns.set_context('notebook') 
sns.set_style("darkgrid")
```

```python
ir = data('iris')
```

## Exercise Section 1
### 1. What does the distribution of petal lengths look like?

```python
ir.head()
```

```python
ir.info()
```

```python
sns.displot(data = ir, x = 'Petal.Length', kde=True)
plt.show()
```

#### This looks a bit like there it is a bimodal distribution

### 2. Is there a correlation between petal length and petal width?

```python
sns.scatterplot(data = ir, x = 'Petal.Length', y='Petal.Width', hue= 'Species')
plt.show()
```

#### There is definitely a strong correlation between petal length and petal widths

### 3. Would it be reasonable to predict species based on sepal width and sepal length? For this, you'll visualize two numeric columns through the lense of a categorical column.

```python
sns.relplot(data = ir, x = 'Sepal.Width', y = 'Sepal.Length', hue = 'Species')
plt.show()
```

```python
ir.Species.unique()
```

#### It appears to me that sepal length could be used to predict species a bit. It's not going to be good though. Sepal width is not going to be a good predictor between the three different species.

### 4. Which features would be best used to predict species?

```python
sns.pairplot(ir, hue = 'Species')
plt.show()
```

#### I think pairplot, coloring Species gives us the best capability to determine if any features between length, or width of petals or sepals can predict species. Petal length with petal width looks to give us a pretty great ability to predict species.
