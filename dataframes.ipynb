### How many animals are hoppers that are above the median speed? What percentage is this?

```python
mammals[(mammals.hoppers == True) & (mammals.speed > mammals.speed.median())]
```

```python
print('The percent of hoppers above the median speed is',round(len(mammals[(mammals.hoppers == True) &
            (mammals.speed > mammals.speed.median())]) / len(mammals) * 100, 2),'%')
```
